```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
athletes = pd.read_csv('data/summerOly_athletes.csv')
hosts = pd.read_csv('data/summerOly_hosts.csv')
medal_counts = pd.read_csv('data/summerOly_medal_counts.csv')
programs = pd.read_csv('data/summerOly_programs.csv')
```

```python
us_medals = medal_counts.where(medal_counts['NOC'] == 'United States').dropna()
plt.scatter(us_medals['Year'], us_medals['Total'])
```

```python
medal_counts_total = medal_counts.groupby(['NOC']).sum()['Total']
medal_counts_total
```
