```python
import pandas as pd
```

```python
sim_data = pd.read_csv('data/sim_full_medal_data.csv').sort_values(by=['Country Code', 'Year']).reset_index(drop=True)
sim_data
```

```python
countries = sim_data['Country Code'].unique()

sim_data.loc[2605, 'Total Male Participants'] = 20
sim_data.loc[2605, 'Total Female Participants'] = 15
sim_data.loc[2606, 'Total Male Participants'] = 23
sim_data.loc[2606, 'Total Female Participants'] = 16

for country_name in countries:
    country = sim_data[sim_data['Country Code'] == country_name]

    if country.iloc[-1, 7] != 2028:
        continue

    try:
        country.iloc[-1, 12:17] = country.iloc[-3:-1, 12:17].mean().astype('int64')
    except Exception as e:
        display(country)
        raise

    sim_data[sim_data['Country Code'] == country_name] = country

sim_data
```

```python
sim_data.to_csv('data/sim_full_medal_data.csv')
```
